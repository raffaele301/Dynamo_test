```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#I checked on the website and there are 57 page so I run a cycle for each page to get the shoes catalog
columns=['brand','product','price','sex','product_type']
farfetch_shoes_catalog = pd.DataFrame( columns=columns)

for page in range(1,58):
    url  = requests.get("https://www.farfetch.com/uk/shopping/men/shoes-2/items.aspx?page="+str(page))
    data = url.text
    soup = BeautifulSoup(data)
    price = soup.findAll('span', {'class':'listing-item-content-price'})
    brand = soup.findAll('h5', {'class':'listing-item-content-brand force-ltr'})
    name = soup.findAll('p', {'class':'listing-item-content-description'})
    for i in range(len(price)):
        farfetch_shoes_catalog=farfetch_shoes_catalog.append({'brand':brand[i].text,'product':name[i].text,'price':price[i].text[1:], 'sex':'male','product_type':'shoes'}, ignore_index=True)
    print("Page "+ str(page) +" Loaded")
```

##### To extract the Farfetch sale catalog I have seen there are 132 page so I loop through each page to get all the info in sale except the stock .the stock info are saved inside a script as text, so I had to slice the text

```python
columns=['original_price','discount','brand', 'description','sale_price','stock']
farfetch_sale_catalog = pd.DataFrame( columns=columns)

for page in range(1,133):
    url  = requests.get("https://www.farfetch.com/uk/shopping/men/sale/all/items.aspx?page="+ str(page)+"&view=180")
    data = url.text
    soup = BeautifulSoup(data)
    original_price = soup.findAll('span', {'class':'listing-item-content-price'})
    discount = soup.findAll('span', {'class':'listing-item-percentOff'})
    brand = soup.findAll('h5', {'class':'listing-item-content-brand force-ltr'})
    description = soup.findAll('p', {'class':'listing-item-content-description'})
    sale_price = soup.findAll('span', {'class':'listing-item-content-sale color-red'})
    scripts=soup.findAll('script')
    for script in scripts:
        if script.text.find("window.universal_variable.listing")>0:
            stock_text=script.text
            stock_unit=[]
            stock_text=stock_text[stock_text.find('"items":'):]
            while stock_text.find('{')>0:
                item=stock_text[stock_text.find('{'):stock_text.find('}')]
                if item.find('"stock":')>1:
                    stock_unit.append(item[int(item.find('"stock":'))+8:int(item.find('"colo'))-1])
                else:
                    stock_unit.append(0)
                stock_text=stock_text[int(stock_text.find('}'))+1:]
    for i in range(len(sale_price)):
        farfetch_sale_catalog=farfetch_sale_catalog.append({'original_price':original_price[i].text,'discount':round(100*(int(original_price[i].text[1:].replace(',',''))-int(sale_price[i].text[1:].replace(',','')))/int(original_price[i].text[1:].replace(',','')),0),'brand':brand[i].text,'description':description[i].text,'sale_price':sale_price[i].text[1:],'stock':stock_unit[i]}, ignore_index=True)
    print("Page "+ str(page) +" Loaded")
```

##### Net a porter doesn't have any sale at the moment and for male have a different website, as I have already extract male shoes from Farfetch I will extract the same thing from net a porter, but the best thing will be to set a scraper to extract the full catalog to compare how much they are exposed to woman, man and child, how many brand they have and the level of stock.
```python
columns=['brand','product','price','sex','product_type']
netaporter_shoes_catalog = pd.DataFrame( columns=columns)
for page in range(1,20):
    url  = requests.get("https://www.mrporter.com/en-gb/mens/shoes?pn="+str(page))
    data = url.text
    soup = BeautifulSoup(data)
    price = soup.findAll('span', {'class':'pl-products-item__text pl-products-item__text--price'})
    brand = soup.findAll('span', {'class':'pl-products-item__text--brand'})
    name = soup.findAll('span', {'class':'pl-products-item__text--name'})
    for i in range(len(price)):
        netaporter_shoes_catalog=netaporter_shoes_catalog.append({'brand':brand[i].text,'product':name[i].text,'price':price[i].text[1:], 'sex':'male','product_type':'shoes'}, ignore_index=True)
    print("Page "+ str(page) +" Loaded")
```
